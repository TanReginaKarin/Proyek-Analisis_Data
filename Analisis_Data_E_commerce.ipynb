## Visualization & Explanatory Analysis

```python
# Mengonversi 'order_purchase_timestamp' ke tipe data datetime
order_items_orders_dataset['order_purchase_timestamp'] = pd.to_datetime(order_items_orders_dataset['order_purchase_timestamp'])

# Menetapkan 'order_purchase_timestamp' sebagai indeks
order_items_orders_dataset.set_index('order_purchase_timestamp', inplace=True)

# Menyusun ulang DataFrame per bulan
monthly_orders_df = order_items_orders_dataset.resample(rule='M').agg({
    "order_id": "nunique",
    "price": "sum"
})

# Menyetel ulang indeks dan memformat tanggal
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df['order_purchase_timestamp'] = monthly_orders_df['order_purchase_timestamp'].dt.strftime('%y-%m')

# Menamai ulang kolom
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)

# Menampilkan DataFrame hasilnya
monthly_orders_df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(monthly_orders_df["order_purchase_timestamp"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Total Order per Bulan (2016-2018)", loc="center", fontsize=20)
plt.xlabel("Tanggal Pemesanan")
plt.ylabel("Jumlah Order")
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.show()
```

```python
# menampilkan performa penjualan (total revenue per bulan) dengan line chart
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders_df["order_purchase_timestamp"],
    monthly_orders_df["revenue"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)

plt.title("Total Revenue per Bulan (2016-2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Sort the data first
grouped_data = grouped_data.sort_values(by="Jumlah", ascending=False)

# Take the top and worst 5 categories
top_5_categories = grouped_data.head(5)
worst_5_categories = grouped_data.tail(5)


# Create the plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 6))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Bar plot for the top categories
sns.barplot(x="Jumlah", y="product_category_name_english", data=top_5_categories, palette=colors, ax=ax1)
ax1.set_ylabel(None)
ax1.set_xlabel(None)
ax1.set_title("Best Performing Product", loc="center", fontsize=15)
ax1.tick_params(axis='y', labelsize=12)

# Bar plot for the worst categories
sns.barplot(x="Jumlah", y="product_category_name_english", data=worst_5_categories, palette=colors, ax=ax2)
ax2.set_ylabel(None)
ax2.set_xlabel(None)
ax2.invert_xaxis()
ax2.yaxis.set_label_position("right")
ax2.yaxis.tick_right()
ax2.set_title("Worst Performing Product", loc="center", fontsize=15)
ax2.tick_params(axis='y', labelsize=12)

plt.suptitle("Best Performing Product by Number of Sales (5 Star Rating)", fontsize=20)
plt.show()
```
